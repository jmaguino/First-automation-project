```python
import openpyxl
import os
import pandas as pd
from openpyxl import load_workbook
import numpy
```

```python
daily_totals = pd.read_excel ('SALES LOG - XYZ COMPANY.xlsx')
```

```python
main_total = pd.read_excel ('tabula-END OF DAY REPORT - XYZ COMPANY.xlsx', header=3) 
```

```python
# Get the grand total from sales log
daily_grand_total = daily_totals.loc[daily_totals['TOTAL '].str.contains('GRAND TOTAL', na=False, case=False), 'QUOTE AMT']
```

```python
# Get the grand total from the end of day report
main_grand_total = main_total.loc[main_total['AMOUNT'].str.contains('GRAND TOTAL', na=False), 'AMOUNT'].str.split('$').str[1].str.replace(',', '').astype(float).values[0]
```

```python
main_total
```

```python
# They happen to be equal for your example
daily_grand_total == main_grand_total
```

```python
main_total
```

```python
# Forward fill blank POs
main_total['PO #'] = main_total['PO #'].ffill()
```

```python
# Filter out NAs
main_total = main_total.loc[~main_total['AMOUNT'].isna()].copy()
```

```python
main_total.columns
```

```python
# Fill subtotal NAs
main_total.loc[main_total['CUSTOMER ID'].isna(), 'CUSTOMER ID'] = 'Subtotal'
main_total.loc[main_total['CUSTOMER ID'].isna(), 'DATE OF SALERECEIPT #'] = 'Subtotal'
main_total.loc[main_total['CUSTOMER ID'].isna(), 'SPEED TYPE'] = 'Subtotal'
```

```python
# Filter the messy daily totals with the cleaner POs
totals_filter = main_total['PO #'].unique()
```

```python
# Use numbers only
totals_filter = [i for i in totals_filter if isinstance(i, int) is True]
```

```python
main_total.columns
```

```python
# View the main totals again with the filter
main_total_viz = main_total.loc[main_total['PO #'].isin(totals_filter)][:40].copy()
main_total_viz['AMOUNT'] = main_total_viz['AMOUNT'].astype(float)
main_total_viz.loc[main_total_viz['PO #'].isin(totals_filter)].style.bar(subset=['AMOUNT'], align='mid', color=['#d65f5f', '#5fba7d']).highlight_null(null_color='red')
```

```python
# View the PO Breakdown
daily_totals.loc[daily_totals['PO #'].isin(totals_filter)].style.bar(subset=['SALE AMT', 'TOTAL '], align='mid', color=['#d65f5f', '#5fba7d']).highlight_null(null_color='red')
```
